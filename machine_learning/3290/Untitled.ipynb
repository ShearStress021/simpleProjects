```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris1 = datasets.load_iris()
```

```python
print(iris1.data)
```

```python
clustering = KMeans(n_clusters=3)
```

```python
clustering.fit(iris1.data)
```

```python
labels = clustering.predict(iris1.data)
```

```python
print(labels)
```

```python
centroids = clustering.cluster_centers_
```

```python
centroids
```

```python
plt.scatter(iris1.data[:,0], iris1.data[:,2],c=labels)
```

```python
plt.scatter(iris1.data[:,0], iris1.data[:,2],c=labels)
plt.scatter(centroids[:,0],centroids[:,2],marker='x', s=150, c='crimson')
plt.show()
```
