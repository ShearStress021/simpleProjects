```python
data = pd.read_csv('data/sp500_companies.csv')
data
```

```python
data['Symbol']
```

```python
list(data['Symbol'])
```

```python
ll = list(data['Symbol'])
```

```python
ll
```

```python
def extract_data(symbol):
    df = yf.download(symbol, start="2023-01-01", end="2026-01-01")
    df.columns = df.columns.droplevel(1)
    df = df.reset_index()
    df['name'] = symbol
    return df    
```

```python
all_symbol = []
for ele in ll:
    data = extract_data(ele)
    all_symbol.append(data)
```

```python
final_df = pd.concat(all_symbol, ignore_index=True)
```

```python
final_df.to_csv("final.csv")
```
